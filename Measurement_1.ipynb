```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def create_plot(data, data_1550, data_850):
    
    x = data['BC']
    d4 = data['C4'][40:100]
    d5 = data['C5'][100:140]
    d8 = data['C8'][50:110]

    
    d4_850 = data_850['C4'][40:100]
    d5_850 = data_850['C5'][100:140]
    d8_850 = data_850['C8'][50:110]

    d4_1550 = data_1550['C4'][40:100]
    d5_1550 = data_1550['C5'][100:140]
    d8_1550 = data_1550['C8'][50:110]

    # Create subplots (2 rows, 2 columns), with the third plot spanning across the second row
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))
    
    axs[0, 0].plot(x[40:100], d4, label='Detector 4', color='b', linestyle='-')
    axs[0, 0].plot(x[40:100], d4_850, label='Detector 4', color='g', linestyle='-')
    axs[0, 0].plot(x[40:100], d4_1550, label='Detector 4', color='r', linestyle='-')
    axs[0, 0].set_title('Detector 4')
    axs[0, 0].grid(True)
    
    axs[0, 1].plot(x[100:140], d5, label='Detector 5', color='b', linestyle='-')
    axs[0, 1].plot(x[100:140], d5_850, label='Detector 5', color='g', linestyle='-')
    axs[0, 1].plot(x[100:140], d5_1550, label='Detector 5', color='r', linestyle='-')
    axs[0, 1].set_title('Detector 5')
    axs[0, 1].grid(True)
    
    axs[1, 0].plot(x[50:110], d8, label='Detector 8', color='b', linestyle='-')
    axs[1, 0].plot(x[50:110], d8_850, label='Detector 8', color='g', linestyle='-')
    axs[1, 0].plot(x[50:110], d8_1550, label='Detector 8', color='r', linestyle='-')
    axs[1, 0].set_title('Detector 8')
    axs[1, 0].grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.xlabel('Time (ns)', fontsize=12)
    plt.ylabel('Voltage (mV)', fontsize=12)
    # Remove the empty subplot (bottom-right)
    fig.delaxes(axs[1, 1])
    
    # Adjust layout
    plt.tight_layout()
    # Show the plot
    plt.show()
```

```python
file_path = "ex1/nophotons.txt"
file_path_850 = "ex1/850nm.txt"
file_path_1550 = "ex1/1550nm.txt"

data = pd.read_csv(file_path, skiprows=9, delimiter="\t")
data_850 = pd.read_csv(file_path_850, skiprows=9, delimiter="\t")
data_1550 = pd.read_csv(file_path_1550, skiprows=9, delimiter="\t")

create_plot(data, data_1550, data_850)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for 3 different graphs


x = data['BC']
d4 = data['C4'][60:100]
d5 = data['C5'][120:135]
d8 = data['C8'][80:105]


d4_850 = data_850['C4'][60:100]
d5_850 = data_850['C5'][120:135]
d8_850 = data_850['C8'][80:105]

d4_1550 = data_1550['C4'][60:100]
d5_1550 = data_1550['C5'][120:135]
d8_1550 = data_1550['C8'][80:105]
# Create a figure with 3 subplots in a row
fig, axes = plt.subplots(1, 3, figsize=(15, 4))  # 1 row, 3 columns

# Plot 1: Sine wave
axes[0].plot(x[60:100], d4, label='No Photons', color='b')
axes[0].plot(x[60:100], d4_850, label='850nm', color='g')
axes[0].plot(x[60:100], d4_1550, label='1550nm', color='r')
axes[0].axvline(x=21.75, color='blue', linestyle='--', label=f'I = 21.75')

axes[0].set_xlabel('Bias current (uA)', fontsize=12)
axes[0].set_ylabel('Counts', fontsize=12)
axes[0].set_title('Detector 4 (60nm, 10um)', fontsize=14)
axes[0].grid(True, which='both', linestyle='--', linewidth=0.5)
axes[0].legend(loc='best')


axes[1].plot(x[80:105], d8, label='No Photons', color='b')
axes[1].plot(x[80:105], d8_850, label='850nm', color='g')
axes[1].plot(x[80:105], d8_1550, label='1550nm', color='r')
axes[1].axvline(x=25.25, color='blue', linestyle='--', label=f'I = 25.25')
axes[1].set_xlabel('Bias current (uA)', fontsize=12)
axes[1].set_ylabel('Counts', fontsize=12)
axes[1].set_title('Detector 8 (60nm, 500um)', fontsize=14)
axes[1].grid(True, which='both', linestyle='--', linewidth=0.5)
axes[1].legend(loc='best')


axes[2].plot(x[120:135], d5, label='No Photons', color='b')
axes[2].plot(x[120:135], d5_850, label='850nm', color='g')
axes[2].plot(x[120:135], d5_1550, label='1550nm', color='r')
axes[2].axvline(x=32, color='blue', linestyle='--', label=f'I = 32')
axes[2].set_xlabel('Bias current (uA)', fontsize=12)
axes[2].set_ylabel('Counts', fontsize=12)
axes[2].set_title('Detector 5 (100nm, 500um)', fontsize=14)
axes[2].grid(True, which='both', linestyle='--', linewidth=0.5)
axes[2].legend(loc='best')


# Adjust layout so titles and labels don't overlap
plt.tight_layout()
plt.savefig('count_vs_current.png', dpi=300) 
# Show the plot
plt.show()
```
